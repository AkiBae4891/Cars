```
import pandas as pd
import numpy as np
```

```
# read in csv
cars = pd.read_csv('cars.csv')
print(cars.shape)
cars.head(3)
```

```
# feature engineering
cars.columns = ['Make', 'Model', 'Year', 'Fuel', 'HP', 'Cylinders', 'Transmission', 'Drivetrain', 'Doors', 'Category', 'Size', 'Style', 'Highway_MPG', 'City_MPG', 'Popularity', 'MSRP']
```

```
cars.head(3)
```

```
#find how many NA
cars.isnull().sum()
```

```
# fillna() 
# cars['Fuel'].fillna(cars['Fuel'].mode()[0], inplace=True)
# cars['HP'].fillna(cars['HP'].mean()[0], inplace=True)
# cars['Cylinders'].fillna(method='ffill', inplace=True)
# cars['Doors'].fillna(cars['Doors'].mode()[0], inplace=True)
# cars['Category'].fillna(method='ffill', inplace=True)
cars.fillna(method='ffill', inplace=True)
```

```
# got all 0's
cars.isnull().sum()
```

```
cars.tail(2)
```

```
# print only rows that have 8 Cylinders 
cars.loc[cars['Cylinders'] == 8]
```

```
# print cars that are v8 and coupes 
 
cars.loc[(cars['Cylinders'] == 8) & (cars['Style'] == 'Coupe')]
```

```
cars['Combined_MPG'] = (cars['Highway_MPG'] + cars['City_MPG']) // 2
```

```
cars['City_MPG'] = cars['City_MPG'].astype('float')
cars['Highway_MPG'] = cars['Highway_MPG'].astype('float')
cars['Combined_MPG'] = cars['Combined_MPG'].astype('float')
```
